```python
class Animal:
    def __init__(self , name , age , color):
        self.name = name
        self.age = age
        self.color = color

    def eat(self):
        print(f"{self.name} is eating")

    def sleep(self):
        print(f"{self.name} is sleeping")

    def run(self):
        print(f"{self.name} is running")
```

```python
a1 = Animal("Dog" , 5 , "Brown")
a2 = Animal("Cat" , 3 , "Black")

a1.eat()
a2.sleep()
```

## inheritance in Python
Inheritance allows one class (child/subclass) to inherit properties and methods from another class (parent/superclass).

Why use it?

* Reuse code

* Extend functionality

* Create relationships between classes

```python
## Parent Class
class Animal :
    def __init__(self, name, species) :
        print("Calling from Animal Class")
        self.name = name
        self.species = species
    def make_sound(self, sound) :
        return f"{self.name} the {self.species} says {sound}" 
```

```python
# class Dog(Animal):
#     def __init__(self, name, breed):  # Dog class intializer 
#         super().__init__(name, "Dog")
#         self.breed = breed

#     def fetch(self):
#         return f"{self.name} is fetching the ball!"
```

```python
# Child Class
class Dog(Animal) :
    def __init__(self, name, breed) :
        print("Initializing Dog class")
        # print(f"{name} is a {breed}")
        super().__init__(name, "Dog")  # parent class initializer
        self.breed = breed
    def make_sound(self) :  # polymorphism
        return super().make_sound("Woof")
```

```python
dog=Dog("Dog1", "Labrador")
dog.species
dog.make_sound()
```

```python
class Vechicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start_engine(self):
        return f"{self.make} {self.model} engine started."

class Car(Vechicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)
        self.num_doors = num_doors

    def car_info(self):
        return f"{self.make} {self.model} with {self.num_doors} doors."    

    def start_engine(self):
        return f"{self.make} {self.model} car engine started with a roar!"

class Motorcycle(Vechicle):
    def __init__(self, make, model, has_sidecar):
        super().__init__(make, model)
        self.has_sidecar = has_sidecar

    def start_engine(self):
        return f"{self.make} {self.model} motorcycle engine started with a vroom!"        
    
    def motor_update(self):
        return f"{self.make} {self.model} motorcycle has {'a sidecar' if self.has_sidecar else 'no sidecar'}."
```

```python

dog = Dog("Buddy", "Golden Retriever")
print(dog.make_sound())
print(f"{dog.name} is a {dog.breed} breed.")
print(dog.species)
```

```python
class Cat(Animal) :
    def __init__(self, name, color) :
        super().__init__(name, "Cat")
        self.color = color
    def make_sound(self) :
        return super().make_sound("Meow")
    
    def purr(self) :
        return f"{self.name} the {self.color} cat is purring."
```

```python
cat=Cat("Whiskers", "Tabby")
print(cat.make_sound())
print(cat.purr())
print(cat.species)
```

# Types of Inheritance in Python

## Single Inheritance

```python
class A: 
    def method_a(self) :
        return "Method A from class A"

class B(A):
    pass    
```

```python
# single inheritance example

class Vehicle:
    def move(self):
        return "The vehicle is moving"
    
class Car(Vehicle):
    def move(self):
        return "The car is driving"
    
car=Car()
print(car.move())    
             
```

# Multilevel Inheritance

```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass
```

```python
# Multiple Inheritance
class A:
    def method_a(self):
        return "Method A from class A"
class B:
    def method_b(self):
        return "Method B from class B"
class C(A, B):

    def method_c(self):
        return "Method C from class C"

c = C()
print(c.method_a())
print(c.method_b())
print(c.method_c())        
```

# Hierarchical Inheritance

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass
```

```python
# Herarchical Inheritance
class A:
    def method_a(self):
        return "Method A from class A"
class B(A):
    def method_b(self):
        return "Method B from class B"
class C(A):
    def method_c(self):
        return "Method C from class C"  
b = B()
c = C()
print(b.method_a())
print(b.method_b())
print(c.method_a())
print(c.method_c())
```
